```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split 
x , y = make_classification(n_samples=1000,n_classes=2,n_features=20,random_state=1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
LRC = LogisticRegression()
DTC = DecisionTreeClassifier()
svc = SVC()
Gnb = GaussianNB()
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
ensembel_model = VotingClassifier(estimators=[('logistic_reg', LRC),('decision_tree',DTC),('svm_classifier',SVC()),('gausian_nv',Gnb)] , voting= 'hard') 
```

```python
ensembel_model
```

```python
ensembel_model.fit(x_train,y_train)
```

```python
ypred  = ensembel_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test , ypred)*100
```
