```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split 
x , y = make_regression(n_samples=1000,n_features=10 , noise= 0.1 , random_state=1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
LRC = LinearRegression()
DTR = DecisionTreeRegressor()
SVR = SVR()
```

```python
from sklearn.ensemble import VotingRegressor
ensembel_model = VotingRegressor(estimators=[('logistic_reg', LRC),('decision_tree',DTR),('svm_classifier',SVR)] ) 
ensembel_model
```

```python
ensembel_model.fit(x_train,y_train)
ypred  = ensembel_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,ypred)*100
```
