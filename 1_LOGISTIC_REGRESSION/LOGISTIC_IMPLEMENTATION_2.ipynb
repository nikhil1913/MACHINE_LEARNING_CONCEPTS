```python
# by default prediction class is defined by cutoff = 0.5
# but if want to define cutoff he use precision recall accuracy trade off
```

```python
from sklearn.metrics import precision_score , recall_score , accuracy_score

threshold = np.linspace(0,1,100)
precision = []
recall = []
accuraccy = []

for i in threshold:
    y_pred_threshold = (y_pred_proba >= i).astype(int)
    p = precision_score(y_test , y_pred_threshold)
    r = recall_score(y_test , y_pred_threshold)
    a = accuracy_score(y_test , y_pred_threshold)
    precision.append(p)
    recall.append(r)
    accuraccy.append(a)

```

```python
plt.plot(threshold , precision , label = "precision")
plt.plot(threshold , recall , label = "recall")
plt.plot(threshold , accuraccy , label = "accuracy")
plt.xlabel("threshold")
plt.ylabel("score")
plt.legend()
plt.title("precision recall accuracy trade off plot")
plt.grid(True)
plt.show()
```
