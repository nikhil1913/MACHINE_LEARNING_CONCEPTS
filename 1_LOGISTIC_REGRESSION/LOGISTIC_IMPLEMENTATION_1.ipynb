```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
data
```

```python
df = pd.DataFrame(data.data , columns = data.feature_names)
```

```python
df
```

```python
df['target'] = data.target
```

```python
df
```

```python
df['target'].value_counts()
```

```python
# we want to understand binary classification here so dropping the tagert with value 2

df_final = df[df['target']!=2]
```

```python
df_final
```

```python
df_final.info()
```

```python
df.isna().sum()
```

```python
from sklearn.model_selection import  train_test_split
x = df_final.iloc[:,:-1]
y = df_final.iloc[:,-1]
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
classifier
```

```python
classifier.fit(x_train , y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
# by default cuttoff is 0.5
classifier.predict_proba(x_test)
```

```python
from sklearn.metrics import confusion_matrix , classification_report , accuracy_score , roc_curve , auc
```

```python
print(confusion_matrix(y_test , y_pred))
```

```python
acc = accuracy_score(y_test , y_pred)
acc
```

```python
print(classification_report(y_test , y_pred))
```

```python
classifier.predict_proba(x_test)[: ,-1] # this is the predicted probability for 1 class
```

```python
y_predict_proba = classifier.predict_proba(x_test)[: ,-1] 
```

```python
print(roc_curve(y_test , y_predict_proba)) # it will return a TPR FPR with different cuttoff of probability
```

```python
fpr , tpr , threshold = roc_curve(y_test , y_predict_proba)
```
